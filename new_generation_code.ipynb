```python
# Extraido de: https://stackoverflow.com/questions/19798112/convert-pandas-dataframe-to-a-nested-dict
def recur_dictify(frame):
    if len(frame.columns) == 1:
        grouped = frame.groupby(frame.columns[0])
        d = {k: [] for k, g in grouped}
        return frame
    
    grouped = frame.groupby(frame.columns[0])
    d = {k: recur_dictify(g.iloc[:,1:]) for k, g in grouped}
    return d


df_tree = estructura[list(estructura.columns)[:5]].copy(deep=True)
dict_tree = recur_dictify(df_tree)

# with open('tree.pickle', 'wb') as handle:
#     pickle.dump(dict_tree, handle, protocol=pickle.HIGHEST_PROTOCOL)

# display(df_tree)
```
